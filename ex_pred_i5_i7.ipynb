Neural Net

```
from keras.models import Sequential
from keras.layers import Dense
```

```
scaler = MinMaxScaler()
scaler.fit(X)
X= scaler.transform(X)

sc_y= MinMaxScaler()
sc_y.fit(y)
y= sc_y.transform(y)
```

```
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.35, random_state=42)

model = Sequential()
model.add(Dense(7, input_dim=10, activation='relu'))
model.add(Dense(4, activation='relu'))
#model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss='mean_squared_error', optimizer='adam')


model.fit(X_tr, y_tr, epochs=30, batch_size=30, verbose=1)
predictions = model.predict(X_te)
print(mean_squared_error( y_te,predictions))
pred_nn= sc_y.inverse_transform(predictions)
y_te= sc_y.inverse_transform(y_te)

for i in range(len(y_te)):
  print('True: ',y_te[i][0],'Pred: ',pred_nn[i], 'Diff= ',y_te[i][0]- pred_nn[i] )
```

```
mean_squared_error(pred_nn, y_te)
```

```
r2_score(pred_nn, y_te)
```
